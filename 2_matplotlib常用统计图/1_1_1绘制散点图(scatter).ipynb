```python
from matplotlib import pyplot as plt
from matplotlib import rc
```

```python
%matplotlib inline
```

```python
y_3 = [11,17,16,11,12,11,12,6,6,7,8,9,12,15,14,17,18,21,16,17,20,14,15,15,15,19,21,22,22,22,23]
y_10 = [26,26,28,19,21,17,16,19,18,20,20,19,22,23,17,20,21,20,22,15,11,15,5,13,17,10,11,13,12,13,6]
```

```python
#设置中文字体(windows和linux设置字体方式)
font = {'family' : 'MicroSoft YaHei',
        'weight' : 'bold'}
rc('font',**font)
```

```python
x_3 = range(1,32)
x_10 = range(51,82)
print(len(x_3),len(y_3))
```

```python
plt.figure(figsize = (20,10),dpi = 80)
plt.scatter(x_3,y_3,label = '3月份')  #绘制散点图
plt.scatter(x_10,y_10,label = '10月份')  #绘制散点图

#调整x轴刻度
_x = list(x_3) + list(x_10)
_xtick_label = ['3月{}日'.format(i) for i in x_3]
_xtick_label += ['10月{}日'.format(i-50) for i in x_10]
plt.xticks(_x[::3],_xtick_label[::3],rotation = 325)

#添加描述信息
plt.xlabel('月份')
plt.ylabel('温度')
plt.title('3月份和10月份温度对比')

#添加图例
plt.legend(loc = 2)


plt.show()
```
