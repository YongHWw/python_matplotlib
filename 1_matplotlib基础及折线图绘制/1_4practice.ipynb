```python
from matplotlib import pyplot as plt
from matplotlib import rc
```

```python
%matplotlib inline
```

```python
#设置中文字体(windows和linux设置字体方式)
font = {'family' : 'MicroSoft YaHei',
        'weight' : 'bold'}
rc('font',**font)
```

```python
y = [1,0,1,1,2,4,3,2,3,4,4,5,6,5,3,3,1,1,1,2]
x = [i for i in range(11,31)]
print(x)
```

```python
plt.figure(figsize = (20,10),dpi = 80)
plt.plot(x,y)

x_label = ['{}岁'.format(i) for i in x]
plt.xticks(x,x_label)
plt.xlabel('年龄（岁）')
plt.ylabel('女朋友数量')
plt.title('11岁到30岁，每年交的女（男）朋友数量。')

#绘制网格
plt.grid(alpha=0.4) #alpha设置网格透明度

plt.show()
```
