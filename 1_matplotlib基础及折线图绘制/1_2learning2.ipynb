```python
from matplotlib import pyplot as plt
from matplotlib import font_manager,rc
import random
```

```python
%matplotlib inline
```

```python
#设置中文字体(windows和linux设置字体方式)
font = {'family' : 'MicroSoft YaHei',
        'weight' : 'bold'}
rc('font',**font)
```

#### 模拟显示一天10点到12点每分钟的气温 

```python
x = range(0,120)
y = [random.randint(20,35) for i in range(120)]
```

```python
plt.figure(figsize=(20,10),dpi=80)
plt.plot(x,y)

#调整x轴刻度
_xticks_label = ['10点{}分'.format(i) for i in range(60)]
_xticks_label += ['11点{}分'.format(i) for i in range(60)]
#xticks(a,b)将原本刻度是a的刻度名字改为b
plt.xticks(list(x)[::3],_xticks_label[::3],rotation=325)  #rotation刻度旋转的度数

#添加标题等描述信息
plt.xlabel('时间')
plt.ylabel('温度 单位（℃）')
plt.title('10点到12点每分钟的气温变化情况')

plt.show()
```
