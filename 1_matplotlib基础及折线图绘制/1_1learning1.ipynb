```python
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
```

```python
x = range(2,26,2)
y = [15,13,14.5,17,20,25,26,26,24,22,18,15]
```

##### 绘图

```python
# 设置图片大小
# figsize = (float,float) 传入一个元组参数，分别设置图片宽和高
# dpi设置图片清晰度
plt.figure(figsize=(20,8),dpi=80)
plt.plot(x,y)

#设置x轴刻度(y轴（yticks）)
xticks_labels = [i/2 for i in range(4,49)]
plt.xticks(xticks_labels[::3])

#保存图片
#plt.savefig("./learning1.png")

plt.show()
```
